# **Perbedaan Resort Hotel dan City Hotel:**

City hotel dan resort hotel adalah dua jenis akomodasi yang berbeda, dan mereka menawarkan pengalaman yang berbeda pula kepada tamu mereka. Berikut adalah perbedaan utama antara keduanya:

**Lokasi:**

City hotel biasanya terletak di pusat kota atau daerah perkotaan yang ramai. Mereka sering kali berdekatan dengan pusat bisnis, pusat perbelanjaan, tempat-tempat hiburan, dan atraksi wisata kota.
Resort hotel, di sisi lain, biasanya terletak di daerah yang lebih terpencil atau pedesaan, seperti pantai, pegunungan, atau pulau. Mereka cenderung menawarkan akses mudah ke alam dan aktivitas luar ruangan.

**Fasilitas dan Layanan:**

City hotel cenderung menawarkan fasilitas dan layanan yang lebih fokus pada kebutuhan bisnis dan perjalanan bisnis. Contohnya termasuk pusat bisnis, ruang rapat, akses internet cepat, layanan laundry cepat, dan restoran-restoran bisnis.
Resort hotel cenderung menawarkan fasilitas dan layanan yang lebih bersifat rekreasi dan liburan. Ini bisa termasuk kolam renang, spa, pusat kebugaran, klub anak-anak, kegiatan olahraga air, dan tur keliling alam.

**Atmosfer:**

Atmosfer di city hotel lebih sibuk dan terkadang lebih formal. Mereka sering menjadi tempat yang cocok untuk pelancong bisnis atau wisatawan yang ingin menjelajahi kehidupan kota.
Resort hotel biasanya memiliki atmosfer yang lebih santai dan tenang. Mereka sering menjadi pilihan yang baik untuk liburan keluarga atau pasangan yang ingin bersantai dan menikmati lingkungan alam yang indah.

**Harga:**

City hotel cenderung memiliki tarif yang lebih tinggi, terutama di kota-kota besar dan populer.
Resort hotel juga dapat memiliki tarif tinggi, tetapi harga mereka bisa bervariasi tergantung pada musim dan lokasi. Beberapa resort juga menawarkan paket liburan all-inclusive yang mencakup makanan dan aktivitas tertentu.

# **Tipe-tipe customer:**

**Contract**: Ini adalah tamu atau kelompok tamu yang memesan kamar hotel sebagai bagian dari kontrak jangka panjang, seperti perusahaan yang memesan kamar untuk karyawan mereka selama periode waktu tertentu atau perjanjian khusus dengan agen perjalanan.

**Group**: Jenis tamu ini adalah kelompok besar yang memesan sejumlah kamar hotel secara bersamaan, biasanya untuk keperluan acara atau pertemuan, seperti konferensi, pernikahan, atau tur wisata. Grup ini biasanya memiliki tarif khusus atau diskon kelompok.

**Transient**: Transient mengacu pada tamu yang melakukan reservasi secara individual, bukan sebagai bagian dari kontrak atau kelompok. Mereka bisa menjadi turis, pelancong bisnis, atau tamu lain yang memesan kamar secara langsung atau melalui platform pemesanan online untuk menginap dalam waktu singkat.

**Transient-Party**: Ini adalah tamu yang menginap sebagai bagian dari kelompok namun bukan dalam kontrak. Contohnya adalah tamu yang bepergian bersama tetapi memesan kamar secara terpisah atau kelompok yang tidak mencakup jumlah kamar yang memenuhi syarat sebagai tamu grup.

**Definisi ADR:**
ADR adalah singkatan dari "Average Daily Rate" (Tarif Harian Rata-rata). Ini adalah metrik yang umum digunakan dalam industri perhotelan untuk mengukur rata-rata pendapatan kamar hotel dalam satu hari tertentu. ADR dihitung dengan membagi total pendapatan kamar yang dihasilkan oleh hotel dalam satu hari dengan jumlah kamar yang terjual pada hari itu. ADR memberikan gambaran tentang harga rata-rata yang dibebankan oleh hotel kepada tamu per malam dan membantu hotel dalam mengelola harga kamar mereka serta mengoptimalkan pendapatan.

# Total Guest Berdasarkan Bulan

```python
import matplotlib.pyplot as plt
import calendar

# Filter data dari Juli 2015 hingga Juni 2017
df_filtered = df[(df['arrival_date'] >= '2015-07-01') & (df['arrival_date'] <= '2017-06-30')]

# Ekstrak tahun dan bulan dari kolom 'arrival_date'
df_filtered['year'] = df_filtered['arrival_date'].dt.year
df_filtered['month_abbr'] = df_filtered['arrival_date'].dt.month.apply(lambda x: calendar.month_abbr[x])

# Buat kolom baru 'month_year' sebagai kombinasi tahun dan singkatan bulan
df_filtered['month_year'] = df_filtered['month_abbr'] + '-' + df_filtered['year'].astype(str)

# Kelompokkan data berdasarkan 'month_year' dan 'hotel', dan hitung total tamu
grouped_data = df_filtered.groupby(['month_year', 'hotel'])['total_guests'].sum().unstack()

# Urutkan DataFrame berdasarkan indeks pengurutan baru
grouped_data_sorted = grouped_data.reindex(sorted(grouped_data.index, key=lambda x: (int(x[-4:]), list(calendar.month_abbr).index(x[:3]))))

# Plot line chart
ax = grouped_data_sorted.plot(kind='line', figsize=(12, 6), marker='o')
plt.title('Total Guest Monthly from Jul15 to Jun17 (City vs. Resort Hotel)')
plt.xlabel('Month-Year')
plt.ylabel('Total Guests')

# Set xticks label
xticks_labels = grouped_data_sorted.index.tolist()
ax.set_xticks(range(len(xticks_labels)))
ax.set_xticklabels(xticks_labels, rotation=45)

plt.legend(title='Hotel Type')
plt.grid(True)
plt.show()
```

# **Total Guest Berdasarkan Hari Per Bulan**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Urutan bulan yang diinginkan
months_order_1 = ['July 2015', 'August 2015', 'September 2015', 'October 2015', 'November 2015', 'December 2015',
                'January 2016', 'February 2016', 'March 2016', 'April 2016', 'May 2016', 'June 2016']

months_order_2 = ['July 2016', 'August 2016', 'September 2016', 'October 2016', 'November 2016', 'December 2016',
                'January 2017', 'February 2017', 'March 2017', 'April 2017', 'May 2017', 'June 2017']

# Filter data dari Juli 2015 hingga Juni 2016
df_filtered_1 = df[(df['arrival_date'] >= '2015-07-01') & (df['arrival_date'] <= '2016-06-30')]

# Filter data dari Juli 2016 hingga Juni 2017
df_filtered_2 = df[(df['arrival_date'] >= '2016-07-01') & (df['arrival_date'] <= '2017-06-30')]

# Buat kolom baru dengan urutan bulan untuk kedua rentang waktu
df_filtered_1['month_order'] = df_filtered_1['arrival_date'].dt.strftime('%B %Y')
df_filtered_2['month_order'] = df_filtered_2['arrival_date'].dt.strftime('%B %Y')

# Buat kolom baru dengan nama hari untuk kedua rentang waktu
df_filtered_1['day_of_week'] = df_filtered_1['arrival_date'].dt.day_name()
df_filtered_2['day_of_week'] = df_filtered_2['arrival_date'].dt.day_name()

# Kelompokkan data untuk kedua rentang waktu berdasarkan urutan bulan dan nama hari, dan hitung rata-rata total tamu
grouped_data_1 = df_filtered_1.groupby(['month_order', 'day_of_week'])['total_guests'].mean().unstack()
grouped_data_2 = df_filtered_2.groupby(['month_order', 'day_of_week'])['total_guests'].mean().unstack()

# Plot line chart untuk rentang waktu pertama
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10))

grouped_data_1.plot(kind='line', ax=ax1, marker='o')
ax1.set_title('Average Daily Guests from Jul15 to Jun16 by Day Name')
ax1.set_xlabel('Month')
ax1.set_ylabel('Average Daily Guests')
ax1.legend(title='Day of Week', loc='upper left', bbox_to_anchor=(1, 1))
ax1.grid(True)
ax1.set_xticks(range(len(months_order_1)))
ax1.set_xticklabels(months_order_1, rotation=45)

# Plot line chart untuk rentang waktu kedua
grouped_data_2.plot(kind='line', ax=ax2, marker='o')
ax2.set_title('Average Daily Guests from Jul16 to Jun17 by Day Name')
ax2.set_xlabel('Month')
ax2.set_ylabel('Average Daily Guests')
ax2.legend(title='Day of Week', loc='upper left', bbox_to_anchor=(1, 1))
ax2.grid(True)
ax2.set_xticks(range(len(months_order_2)))
ax2.set_xticklabels(months_order_2, rotation=45)

plt.tight_layout()
plt.show()
```

Dalam banyak negara, khususnya di belahan bumi utara, musim panas (Juni, Juli, Agustus) sering kali menjadi waktu liburan utama, di mana orang cenderung melakukan perjalanan dan menginap di hotel. Oleh karena itu, bulan Agustus, yang merupakan musim panas di belahan bumi utara, mungkin memiliki jumlah pengunjung terbanyak.

Sementara itu, bulan Desember sering kali menjadi waktu liburan akhir tahun di banyak negara, terutama karena liburan Natal dan Tahun Baru. Banyak orang merencanakan liburan mereka selama periode ini, yang dapat menyebabkan peningkatan kunjungan ke hotel.

Di sisi lain, bulan November dan April mungkin memiliki jumlah pengunjung yang lebih sedikit karena mereka mungkin dianggap sebagai musim peralihan. Bulan November sering kali berada di antara musim panas dan musim gugur, sementara April adalah bulan di antara musim dingin dan musim semi. Selain itu, di beberapa negara, bulan November sering kali tidak memiliki liburan besar seperti yang terjadi di bulan Desember.

Jadi, bisa jadi pembagian musim memengaruhi pola kunjungan pengunjung ke hotel, dengan musim panas dan liburan akhir tahun cenderung memiliki jumlah pengunjung yang lebih tinggi daripada musim-musim lainnya.
